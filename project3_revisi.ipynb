# **I. BUSINESS UNDERSTANDING**

**1.1. Latar Belakang**

Harga rumah di California dipengaruhi oleh berbagai faktor kompleks, seperti lokasi, ukuran properti, kondisi pasar, dan karakteristik lingkungan sekitar. Beberapa faktor utama yang memengaruhi harga rumah antara lain jumlah populasi, jumlah rumah tangga, total ruangan dan kamar, kedekatan dengan pantai, usia bangunan, serta pendapatan rata-rata penduduk setempat. Faktor-faktor ini saling berinteraksi untuk menentukan nilai pasar properti di wilayah tersebut.

**1.2. Permasalahan**

Masalah utama yang dihadapi adalah bagaimana cara **memprediksi harga rumah** secara akurat berdasarkan berbagai faktor yang mempengaruhinya, seperti ukuran rumah, lokasi, pendapatan penduduk, dan kedekatan dengan fasilitas umum atau pantai. Prediksi harga yang akurat sangat penting untuk membantu para pelaku bisnis properti dalam **menetapkan harga jual yang tepat** dan **menghindari kerugian** akibat kesalahan dalam penetapan harga.

**1.3. Tujuan**

Tujuan utama dari model prediksi harga rumah adalah untuk **menurunkan kesalahan prediksi** (mengurangi **MSE** atau **MAPE**) dengan memanfaatkan data seperti **lokasi**, **ukuran rumah**, **pendapatan median**, dan **jumlah rumah tangga**. Dengan meningkatkan akurasi prediksi, model ini dapat membantu pengembang dan agen properti **menetapkan harga jual** yang lebih akurat dan relevan dengan kondisi pasar.

**1.4.Stakeholder yang Terlibat**

Model **prediksi harga rumah** ini akan digunakan oleh berbagai pihak yang terlibat dalam industri properti, antara lain:
1. **Pengembang properti dan investor real estate**: Mereka akan memanfaatkan model ini untuk memprediksi harga rumah di berbagai wilayah berdasarkan faktor-faktor yang ada dalam data, seperti lokasi, ukuran rumah, dan pendapatan rumah tangga.
2. **Agen real estate**: Mereka akan menggunakan model ini untuk menetapkan harga jual rumah yang lebih akurat.
3. **Pemerintah atau pembuat kebijakan**: Mereka juga menggunakan model untuk merencanakan pembangunan infrastruktur yang berdampak pada harga properti.
4. **Calon pembeli rumah**: Calon pembeli akan memanfaatkan model ini untuk mendapatkan gambaran lebih jelas tentang harga yang wajar dan sesuai dengan anggaran mereka di berbagai kawasan.

Model ini juga akan digunakan setiap kali ada permintaan untuk **memperkirakan harga rumah**, terutama ketika ada perubahan **di pasar properti**, misalnya perubahan demografis atau infrastruktur baru.

**1.5. Metric Evaluation**

Model ini akan dievaluasi menggunakan beberapa **metrik evaluasi** untuk memastikan kualitas prediksi harga rumah, dengan **prioritas pada RMSE**. Metrik yang digunakan adalah:

1. **RMSE (Root Mean Squared Error)**: RMSE adalah metrik yang sangat berguna untuk mengukur kesalahan kuadrat rata-rata antara nilai yang diprediksi dan nilai aktual. Metrik ini memiliki prioritas tertinggi karena **memberikan penalti lebih besar pada kesalahan besar**, yang penting dalam konteks prediksi harga rumah, di mana kesalahan besar bisa sangat mempengaruhi keputusan harga dan investasi.
   
2. **MAE (Mean Absolute Error)**: MAE mengukur rata-rata kesalahan absolut antara prediksi dan nilai aktual, tanpa memberi bobot lebih pada kesalahan besar. Ini memberikan gambaran yang lebih sederhana dan konsisten tentang kesalahan model, meskipun kurang sensitif terhadap perbedaan besar.

3. **MAPE (Mean Absolute Percentage Error)**: MAPE mengukur kesalahan dalam bentuk persentase, memberikan gambaran seberapa besar kesalahan prediksi dalam konteks relatif terhadap nilai aktual. Ini membantu untuk melihat kinerja model dalam skala persentase, meskipun bisa lebih sensitif terhadap nilai-nilai kecil.

    **Prioritas Penggunaan Metrik**:
    Model ini lebih memprioritaskan **RMSE** sebagai metrik utama karena pentingnya meminimalkan kesalahan prediksi yang besar pada harga rumah. Pengurangan RMSE akan memastikan bahwa model tidak hanya akurat dalam rata-rata, tetapi juga dapat menghindari kesalahan besar yang dapat mengarah pada keputusan investasi yang merugikan.


# **II. DATA UNDERSTANDING**
Berikut adalah penjelasan tentang setiap kolom dalam dataset **California Housing** dan relevansinya dalam konteks bisnis properti, khususnya dalam memprediksi harga rumah.

1. **longitude dan latitude**
   - **Deskripsi**: Kedua kolom ini menunjukkan posisi geografis rumah berdasarkan **garis bujur** dan **garis lintang**.
   - **Relevansi Bisnis**: Lokasi properti sangat memengaruhi harga. Rumah yang berada di lokasi strategis, seperti dekat pusat kota atau kawasan yang sedang berkembang, akan memiliki nilai lebih tinggi. Dalam analisis prediksi harga rumah, informasi ini sangat berguna untuk mengidentifikasi area yang lebih bernilai.

2. **housing_median_age**
   - **Deskripsi**: Median usia rumah di suatu area.
   - **Relevansi Bisnis**: Usia rumah sering kali menjadi indikator terkait dengan kualitas properti dan kebutuhan renovasi. Rumah yang lebih tua mungkin memerlukan biaya tambahan untuk perbaikan atau pembaruan, yang dapat mempengaruhi harga jualnya. Data ini membantu dalam memprediksi harga berdasarkan potensi perbaikan atau perawatan yang diperlukan.

3. **total_rooms**
   - **Deskripsi**: Total jumlah ruang yang ada di area tersebut.
   - **Relevansi Bisnis**: Ukuran rumah (dalam hal jumlah ruang) langsung berkorelasi dengan harga properti. Rumah dengan lebih banyak ruang biasanya memiliki harga yang lebih tinggi karena lebih mampu menampung lebih banyak penghuni. Data ini membantu model untuk memperkirakan harga rumah berdasarkan ukuran.

4. **total_bedrooms**
   - **Deskripsi**: Jumlah kamar tidur di wilayah tersebut.
   - **Relevansi Bisnis**: Kamar tidur adalah salah satu ukuran penting dalam menentukan kapasitas rumah. Daerah dengan banyak keluarga atau individu yang membutuhkan lebih banyak ruang tidur cenderung memiliki harga rumah yang lebih tinggi. Ini membantu memahami permintaan untuk properti dengan kapasitas lebih besar.

5. **population**
   - **Deskripsi**: Total populasi yang tinggal di daerah tersebut.
   - **Relevansi Bisnis**: Populasi yang lebih tinggi sering kali menunjukkan permintaan yang lebih besar untuk rumah dan properti lainnya. Area dengan populasi besar biasanya memiliki permintaan lebih tinggi untuk perumahan, yang dapat meningkatkan harga rumah. Dalam prediksi harga rumah, jumlah populasi digunakan untuk mengukur potensi pasar dan permintaan properti.

6. **households**
   - **Deskripsi**: Jumlah rumah tangga di area tersebut.
   - **Relevansi Bisnis**: Semakin banyak rumah tangga, semakin banyak pula permintaan untuk rumah. Data ini penting untuk memahami **kebutuhan perumahan** di daerah tertentu, yang dapat digunakan untuk merencanakan pengembangan properti yang sesuai dengan permintaan pasar.

7. **median_income**
   - **Deskripsi**: Pendapatan median rumah tangga di area tersebut.
   - **Relevansi Bisnis**: **Pendapatan median** adalah salah satu faktor utama yang mempengaruhi daya beli masyarakat. Daerah dengan pendapatan yang lebih tinggi biasanya memiliki permintaan untuk rumah dengan harga lebih tinggi. Dalam memprediksi harga rumah, data ini membantu untuk menyesuaikan harga dengan kemampuan beli masyarakat setempat.

8. **ocean_proximity**
   - **Deskripsi**: Kedekatan dengan laut, apakah rumah tersebut dekat dengan pantai atau tidak.
   - **Relevansi Bisnis**: Lokasi properti yang dekat dengan laut cenderung memiliki harga yang lebih tinggi karena dianggap lebih diinginkan. Dalam bisnis properti, mengetahui kedekatan dengan laut dapat membantu dalam **segmentation pasar** dan menentukan harga yang sesuai untuk area tersebut.

9. **median_house_value**
   - **Deskripsi**: Nilai median rumah di wilayah tersebut (ini adalah target yang ingin diprediksi).
   - **Relevansi Bisnis**: **Harga rumah median** adalah indikator penting untuk menilai pasar properti di area tersebut. Data ini memberikan gambaran umum tentang harga properti dan menjadi parameter utama dalam **penentuan harga jual** rumah. Ini adalah tujuan utama dalam model prediksi harga rumah.

**Kolom-kolom dalam dataset tersebut dapat digunakan untuk Memperkirakan Harga Rumah.**

   1. **Lokasi dan Ukuran**: Kolom seperti **longitude**, **latitude**, **total_rooms**, dan **total_bedrooms** sangat penting dalam menentukan ukuran dan kualitas rumah yang berhubungan langsung dengan harga. Lokasi juga berperan besar dalam menentukan daya tarik dan nilai properti.

   2. **Kepadatan dan Kebutuhan Perumahan**: **Population** dan **households** digunakan untuk mengukur **kebutuhan perumahan**. Area dengan populasi dan rumah tangga yang lebih besar kemungkinan besar membutuhkan lebih banyak rumah, yang dapat meningkatkan harga properti.

   3. **Pendapatan dan Daya Beli**: **Median_income** membantu menentukan apakah masyarakat di area tersebut memiliki daya beli yang cukup untuk membeli rumah dengan harga tertentu. Daerah dengan pendapatan tinggi cenderung membeli rumah dengan harga lebih tinggi.

Nilai **52** yang muncul sebanyak **766 kali** dalam kolom **`housing_median_age`** diduga kuat merupakan **hasil clipping** untuk mengatasi data yang lebih tinggi. Nilai ini merupakan **anomali** karena bukan representasi harga rumah yang wajar, melainkan penyesuaian untuk menghindari outlier yang ekstrem. Karena **52** adalah **angka anomali**  maka akan dilakukan tindakan **menerapkan binning untuk kolom `housing_median_age`**, sehingga data menjadi lebih natural dan analisis lebih akurat.

Pengkategorian usia rumah di pasar properti umumnya digunakan untuk menilai nilai dan kondisi properti.

   - **Keuntungan**: Harga lebih terjangkau, kondisi baik, hanya membutuhkan renovasi ringan.
   - **Risiko**: Beberapa bagian rumah perlu pembaruan seperti pengecatan atau peralatan.

3. **21-30 Tahun**:
   - **Keuntungan**: Harga beli lebih rendah, potensi renovasi untuk meningkatkan nilai jual.
   - **Risiko**: Memerlukan renovasi lebih besar, perbaikan struktural dan sistem.

4. **31-40 Tahun**:
   - **Keuntungan**: Harga beli lebih murah, dengan potensi besar setelah renovasi.
   - **Risiko**: Biaya renovasi tinggi, sistem rumah mungkin sudah usang.

5. **41-50 Tahun**:
   - **Keuntungan**: Potensi keuntungan besar jika renovasi tepat, sering berada di lokasi strategis.
   - **Risiko**: Biaya restorasi tinggi, membutuhkan perbaikan struktural besar.

6. **>50 Tahun**:
   - **Keuntungan**: Nilai sejarah atau arsitektur unik, lokasi strategis.
   - **Risiko**: Biaya restorasi sangat tinggi, renovasi membutuhkan izin khusus.

Secara umum, rumah yang lebih tua cenderung lebih murah namun memerlukan biaya renovasi yang besar. Rumah yang lebih baru lebih stabil, namun harga lebih tinggi dan potensi kenaikan nilai lebih terbatas.

```python
for i in range(len(kolom_numerik)) :
    Q3 = df[kolom_numerik[i]].quantile(0.75)
    Q1 = df[kolom_numerik[i]].quantile(0.25)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    min_data = df[kolom_numerik[i]].min()
    max_data = df[kolom_numerik[i]].max()
    n_outlier = len(df[df[kolom_numerik[i]]>upper_bound]) + len(df[df[kolom_numerik[i]]<lower_bound])
    print(f'{i+1}.  Kolom : {kolom_numerik[i]}')
    print(f'    Jumlah Outlier = {n_outlier}')
    print(f'    Upper bound = {upper_bound:.2f} dan',f'Lower bound = {lower_bound:.2f}')
```

Untuk mengatasi masalah ini, akan diterapkan teknik **binning**, yaitu mengelompokkan data dalam interval atau bin tertentu. Dengan binning, nilai ekstrem akan digabungkan ke dalam rentang yang lebih besar, mengurangi dampak outlier. Misalnya, pada `median_income`, nilai yang sangat tinggi akan dikelompokkan dalam bin yang lebih luas, sehingga tidak mempengaruhi analisis secara signifikan.

Binning juga membantu:
- **Mengurangi variabilitas data**: Meminimalkan fluktuasi ekstrem.
- **Mempermudah interpretasi**: Data lebih mudah dipahami tanpa pengaruh nilai ekstrem.
- **Meningkatkan kestabilan model**: Mengurangi efek outlier pada performa model.

Dengan binning, outlier dapat dikelola dengan lebih baik, memastikan analisis dan model lebih stabil.

Jumlah bin yang disarankan menggunakan rumus **Sturges** dapat dihitung dengan rumus berikut:

$$
k = 1 + \log_2(n)
$$

Dimana:
- \( n = 13637 \) adalah jumlah data.

Langkah-langkah perhitungannya adalah sebagai berikut:

1. Hitung nilai logaritma basis 2 dari \( n = 13637 \):

$$
\log_2(13637) = \frac{\log_{10}(13637)}{\log_{10}(2)} = \frac{4.135}{0.3010} \approx 13.74
$$

2. Tambahkan 1 ke hasilnya:

$$
k = 1 + 13.74 = 14.74
$$

Jadi, jumlah bin yang disarankan adalah sekitar **14.74**, yang biasanya dibulatkan menjadi **15 bin**.


Binning mengelompokkan data ke dalam interval yang lebih besar, yang mengurangi pengaruh nilai ekstrem.Dengan demikian, distribusi menjadi lebih stabil dan lebih mudah dianalisis tanpa gangguan dari nilai-nilai ekstrem yang bisa mempengaruhi hasil analisis. **Binning akan diterapkan pada Pipeline untuk menghidari leakage atau bocornya informasi**

## 4.1. MODEL SELECTION

Algoritma machine learning yang akan diterapkan pada tahap pemodelan adalah sebagai berikut.

1. **Linear Regression (LR)**
   - **Deskripsi**: Linear Regression adalah algoritma regresi dasar yang mencoba untuk memodelkan hubungan linier antara fitur (variabel independen) dan target (variabel dependen). Model ini mencari **koefisien** yang meminimalkan **error** antara nilai prediksi dan nilai aktual dengan menggunakan metode **least squares**.
   - **Karakteristik**:
     - Sederhana dan mudah diinterpretasi.
     - Membutuhkan hubungan linier antara fitur dan target.
     - Tidak fleksibel dalam menangani hubungan non-linier.
   - **Kelebihan**:
     - Cepat dan mudah diimplementasikan.
     - Mudah untuk diinterpretasi dan menjelaskan model.
     - Cocok untuk dataset yang memiliki hubungan linier jelas.
   - **Kekurangan**:
     - Tidak mampu menangani **hubungan non-linier** dengan baik.
     - Sensitif terhadap **outlier**, yang bisa memengaruhi akurasi model.
     - Hanya efektif jika fitur berkolinear rendah.


2. **K-Nearest Neighbors Regressor (KNN)**
   - **Deskripsi**: KNN adalah algoritma non-parametrik yang memprediksi nilai target dengan mencari **k tetangga terdekat** dalam ruang fitur. Model ini menganggap bahwa data yang serupa memiliki nilai target yang serupa.
   - **Karakteristik**:
     - Tidak membutuhkan pelatihan eksplisit; bekerja dengan menyimpan seluruh data pelatihan.
     - Sensitif terhadap **skala data**.
   - **Kelebihan**:
     - Mudah dipahami dan diimplementasikan.
     - Tidak memerlukan asumsi linearitas atau distribusi data.
     - Cocok untuk dataset kecil dan menengah.
   - **Kekurangan**:
     - Lambat untuk dataset besar karena perlu menghitung jarak antara semua titik data.
     - Sangat dipengaruhi oleh **outlier** dan **dimensionality curse** (jumlah fitur yang tinggi dapat menurunkan kinerja).
     - Membutuhkan **normalisasi data** untuk performa yang optimal.


3. **Decision Tree Regressor (DT)**
   - **Deskripsi**: Decision Tree Regressor membagi data ke dalam subset berdasarkan fitur untuk memprediksi nilai target. Pohon keputusan terbentuk dengan memilih fitur yang membagi data secara optimal pada setiap level.
   - **Karakteristik**:
     - Model berbentuk pohon yang mudah diinterpretasi.
     - Dapat menangani data numerik dan kategorikal.
   - **Kelebihan**:
     - Mudah dipahami dan diinterpretasi karena berbentuk pohon.
     - Dapat menangani hubungan non-linier antar fitur.
     - Tidak memerlukan **skala data** yang seragam.
   - **Kekurangan**:
     - Rentan terhadap **overfitting** jika pohon terlalu dalam.
     - Kurang akurat pada data dengan banyak variabilitas atau noise.
     - Tidak cocok untuk dataset besar tanpa **pruning** pohon.

4. **Random Forest Regressor (RF)**
   - **Deskripsi**: Random Forest adalah **ensemble method** yang membangun banyak pohon keputusan secara acak dan menggabungkan hasilnya untuk meningkatkan akurasi dan mengurangi overfitting. Setiap pohon dilatih pada subset acak dari data dan fitur.
   - **Karakteristik**:
     - Ensemble dari banyak pohon keputusan.
     - Menggunakan **bagging** untuk meningkatkan kestabilan dan mengurangi overfitting.
   - **Kelebihan**:
     - Lebih stabil dan akurat dibandingkan pohon keputusan tunggal karena menggabungkan prediksi dari banyak pohon.
     - Dapat menangani **data yang hilang** dan **outlier** dengan baik.
     - Dapat menangani **hubungan non-linier** antar fitur.
   - **Kekurangan**:
     - Modelnya lebih kompleks dan sulit diinterpretasi dibandingkan dengan pohon keputusan tunggal.
     - Memerlukan lebih banyak **waktu komputasi** dibandingkan dengan model lainnya, terutama untuk dataset besar.
     - **Memori lebih besar** karena menyimpan banyak pohon.


5. **XGBoost Regressor (XGB)**
   - **Deskripsi**: XGBoost adalah algoritma **gradient boosting** yang menggunakan pendekatan bertahap untuk membangun model. Setiap pohon yang dibangun mencoba untuk mengoreksi kesalahan dari pohon sebelumnya, sehingga menghasilkan model yang sangat kuat dan akurat.
   - **Karakteristik**:
     - Menerapkan **gradient boosting** yang membangun model bertahap.
     - Mendukung **regularisasi** untuk menghindari overfitting.
     - Dikenal dengan **kecepatan** dan **efisiensi**.
   - **Kelebihan**:
     - Sangat akurat dan seringkali mengungguli model lainnya dalam kompetisi machine learning.
     - Mengurangi **overfitting** melalui **regularisasi**.
     - Dapat menangani **data yang hilang** dan **skala data** yang berbeda.
   - **Kekurangan**:
     - Memerlukan **penyetelan parameter** yang lebih kompleks untuk mencapai kinerja optimal.
     - Lebih rumit untuk diinterpretasi karena merupakan model yang lebih kompleks.

---


```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(-model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error',
        error_score='raise'
        )

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(-model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error',
        error_score='raise'
        )

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(-model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

Pada langkah awal pemodelan regresi akan diukur kinerja model berdasarkan tiga metrik evaluasi (RMSE, MAE, MAPE), dan memilih model terbaik berdasarkan hasil evaluasi tersebut.

1. **Definisi Algoritma yang Digunakan**:
   - `lr = LinearRegression()`: Mendefinisikan model regresi linier.
   - `knn = KNeighborsRegressor()`: Mendefinisikan model regresi dengan K-Nearest Neighbors.
   - `dt = DecisionTreeRegressor(random_state=1)`: Mendefinisikan model regresi dengan pohon keputusan.
   - `rf = RandomForestRegressor(random_state=1)`: Mendefinisikan model regresi dengan hutan acak.
   - `xgb = XGBRegressor(random_state=1)`: Mendefinisikan model regresi dengan XGBoost.

2. **Transformasi Data**:
   - Data yang digunakan untuk pemodelan diubah ke dalam skala logaritmik terlebih dahulu (`log_lr`, `log_knn`, `log_dt`, `log_rf`, `log_xgb`) untuk menstabilkan varians dan kemudian di-inverse kembali ke bentuk aslinya untuk interpretasi.
   
3. **Scaling Data**:
   - `scaler = StandardScaler()`: Digunakan untuk melakukan normalisasi atau standarisasi data, mengubah data sehingga memiliki rata-rata 0 dan deviasi standar 1.

4. **Membuat Daftar Model**:
   - `models`: List yang berisi berbagai model yang akan dievaluasi, yaitu model yang sudah ditransformasi dalam bentuk log.

5. **Metrik Evaluasi**:
   - `score_rmse`, `score_mae`, `score_mape`: List yang akan menyimpan hasil evaluasi (RMSE, MAE, MAPE) untuk setiap model.
   - `nilai_mean_rmse`, `nilai_mean_mae`, `nilai_mean_mape`: List untuk menyimpan rata-rata nilai dari masing-masing metrik.
   - `nilai_std_rmse`, `nilai_std_mae`, `nilai_std_mape`: List untuk menyimpan nilai standar deviasi dari setiap metrik.

6. **Proses Evaluasi Model**:
   - Bagian ini melakukan evaluasi model menggunakan cross-validation dengan 5 fold (KFold).
   - Setiap model dievaluasi menggunakan tiga metrik:
     - **RMSE (Root Mean Squared Error)**: Untuk mengukur rata-rata kesalahan kuadrat yang terakumulasi.
     - **MAE (Mean Absolute Error)**: Mengukur rata-rata kesalahan absolut model.
     - **MAPE (Mean Absolute Percentage Error)**: Mengukur rata-rata persentase kesalahan absolut.

   Proses cross-validation dilakukan untuk setiap metrik menggunakan `cross_val_score()` untuk setiap model dan hasilnya disimpan dalam variabel terkait.

## 4.2 HYPERPARAMETER TUNING

```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = list(np.arange(1, 100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10)) # Semakin besar nilainya, semakin konservatif/simpel modelnya


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}

```

Ada beberapa parameter pada **XGBoost** yang akan dioptimalisasi dalam tuning hyperparameter:

1. **max_depth** 
Parameter ini menentukan kedalaman maksimum pohon keputusan. Semakin dalam pohon, semakin banyak pembagian yang dilakukan, dan semakin kompleks model yang terbentuk. Kedalaman yang lebih besar bisa menangkap pola lebih banyak, tetapi juga meningkatkan risiko **overfitting**. Dengan memilih kedalaman yang tepat, kita dapat menghindari model yang terlalu rumit atau terlalu sederhana.

2. **learning_rate**
Learning rate mengontrol ukuran langkah yang diambil pada setiap iterasi pembelajaran. Nilai yang lebih besar memungkinkan model untuk belajar lebih cepat, namun bisa membuatnya melangkah terlalu jauh dan kehilangan pola penting. Sebaliknya, nilai yang lebih kecil akan memperlambat proses pembelajaran, dan sering kali memerlukan lebih banyak pohon (n_estimators) untuk mencapai hasil yang optimal.

3. **n_estimators** 
Jumlah pohon mengatur berapa banyak pohon yang dibangun selama pelatihan model. Setiap pohon belajar dari kesalahan pohon sebelumnya. Semakin banyak pohon yang digunakan, semakin baik model dalam mengoreksi kesalahan, tetapi jika terlalu banyak, model bisa menjadi terlalu rumit dan berisiko **overfitting**.

4. **subsample** 
Subsample adalah proporsi data pelatihan yang digunakan untuk membangun setiap pohon. Menggunakan sebagian data membuat model lebih efisien dan mengurangi risiko **overfitting** dengan cara membuat model lebih sederhana dan tidak terlalu terikat pada data pelatihan yang ada.

5. **gamma**  
Gamma mengontrol apakah pembagian baru akan dilakukan dalam pohon. Semakin besar nilai gamma, semakin konservatif model dalam melakukan pembagian, hanya melakukannya jika pengurangan impurity cukup besar. Ini mencegah model menjadi terlalu kompleks dan memastikan bahwa hanya pembagian yang benar-benar bernilai yang terjadi.

6. **colsample_bytree**  
Parameter ini mengatur seberapa banyak fitur yang digunakan dalam membangun setiap pohon. Dengan memilih subset fitur, model menjadi lebih efisien dan lebih cepat dalam pelatihan, sekaligus mengurangi kemungkinan **overfitting**. Ini memberikan kontrol lebih terhadap kompleksitas model.

7. **reg_alpha**
Reg_alpha adalah parameter regularisasi L1 yang menambahkan penalti pada bobot fitur. Nilai yang lebih besar akan mendorong model untuk lebih selektif dalam memilih fitur yang penting, sehingga model menjadi lebih sederhana dan lebih tahan terhadap **overfitting**.

```python
from sklearn.model_selection import RandomizedSearchCV

# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
random_xgb1 = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 1000,
    cv = crossval,
    n_jobs=-1, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 1  
)
```

Selanjutnya dilakukan **hyperparameter tuning** untuk model **XGBoost** menggunakan **RandomizedSearchCV** dari **scikit-learn**.

1. **Model XGBoost**:
   **XGBoost Regressor** merupakan salah satu algoritma regresi berbasis pohon keputusan yang sangat populer. Model ini dikonfigurasi dengan beberapa parameter dasar, termasuk penetapan nilai acak (random_state) dan pengaturan verbosity (verbosity=0) untuk menonaktifkan output log yang tidak perlu.

2. **Pipeline**:
   Selanjutnya, dibuat sebuah **pipeline** untuk memproses data sebelum digunakan oleh model. Pipeline ini menggabungkan beberapa langkah:
   - **Preprocessing**: Proses awal untuk memanipulasi atau mengubah data menjadi format yang lebih baik untuk model.
   - **Scaler**: Langkah kedua di pipeline ini digunakan untuk melakukan standarisasi atau normalisasi data, memastikan bahwa semua fitur memiliki skala yang sama agar model lebih efektif.
   - **Model**: Setelah data diproses, model **XGBoost** diterapkan sebagai langkah terakhir.

3. **Cross-validation**:
   Untuk mengevaluasi kinerja model secara lebih akurat, kode ini menggunakan **cross-validation** dengan 5 fold. Ini berarti data dibagi menjadi lima bagian, dan model dilatih lima kali, masing-masing dengan satu bagian sebagai data uji dan empat bagian lainnya sebagai data pelatihan. Ini membantu mengurangi kemungkinan bias dan memastikan model bekerja dengan baik pada data yang belum terlihat.

4. **Hyperparameter Tuning dengan RandomizedSearchCV**:
   **RandomizedSearchCV** digunakan untuk mencari kombinasi terbaik dari berbagai **hyperparameter** model **XGBoost**. Hyperparameter ini mencakup nilai-nilai seperti kedalaman pohon, learning rate, jumlah pohon, dan lain-lain. Proses pencarian dilakukan secara acak, dan sebanyak 1000 kombinasi berbeda dari hyperparameter diuji.
   
   - **Metrik Evaluasi**: Tiga metrik digunakan untuk mengevaluasi kinerja model, yaitu **RMSE** (Root Mean Squared Error), **MAE** (Mean Absolute Error), dan **MAPE** (Mean Absolute Percentage Error). Metrik ini digunakan untuk mengukur seberapa baik model memprediksi data yang sebenarnya.
   - **Refit**: Setelah pencarian selesai, model terbaik dipilih berdasarkan metrik **RMSE**.

```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(-model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error',
        error_score='raise'
        )

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(-model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error',
        error_score='raise'
        )

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(-model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
```

### 6.1.3 LIMITASI MODEL

Setelah dilakukan penghilangan outlier pada kolom target **`median_house_value`**, model menjadi lebih terfokus pada prediksi harga rumah yang realistis dan sesuai dengan kategori lokasi masing-masing. Dampak dari limitasi adalah sebagai berikut.

1. **INLAND:**
   - **Limit Prediksi Harga**: Model hanya dapat memprediksi harga rumah dalam rentang **14999 - 255200 USD** untuk kategori **INLAND**.
   - **Konsekuensi**: Jika harga rumah di daerah pedalaman lebih rendah dari 14,999 USD atau lebih tinggi dari 255,200 USD, model tidak akan dapat memberikan prediksi yang akurat atau mungkin tidak dapat memberikan prediksi sama sekali. Ini membatasi aplikasi model hanya untuk harga rumah dalam rentang tersebut.

2. **NEAR BAY:**
   - **Limit Prediksi Harga**: Model hanya dapat memprediksi harga rumah dalam rentang **22500 - 427300 USD** untuk kategori **NEAR BAY**.
   - **Konsekuensi**: Prediksi harga rumah di wilayah dekat teluk yang berada di luar rentang harga ini (baik lebih rendah dari 22,500 USD atau lebih tinggi dari 427,300 USD) tidak akan akurat. Model menjadi terbatas dalam memperkirakan harga rumah di area tersebut, yang mengurangi fleksibilitas model.

3. **NEAR OCEAN:**
   - **Limit Prediksi Harga**: Model hanya dapat memprediksi harga rumah dalam rentang **37500 - 426400 USD** untuk kategori **NEAR OCEAN**.
   - **Konsekuensi**: Untuk rumah yang terletak dekat dengan laut, jika harga rumah berada di luar rentang ini, model tidak akan dapat memberikan prediksi yang tepat. Ini membatasi aplikasinya pada harga rumah yang realistis di wilayah tersebut.

4. **<1H OCEAN:**
   - **Limit Prediksi Harga**: Model hanya dapat memprediksi harga rumah dalam rentang **17500 - 427300 USD** untuk kategori **<1H OCEAN**.
   - **Konsekuensi**: Model menjadi terbatas dalam memprediksi harga rumah yang lebih rendah dari 17,500 USD atau lebih tinggi dari 427,300 USD. Prediksi harga di luar rentang ini mungkin tidak valid atau bahkan tidak tersedia.

5. **ISLAND:**
   - **Tidak Diperuntukkan**: Karena kategori **ISLAND** memiliki hanya dua data yang sangat sedikit dan tidak representatif, model **tidak diperuntukkan** untuk memprediksi harga rumah di kategori ini.

### 6.2.2 PENGGUNAAN GridSearchCV PADA HYPERPARAMETER TUNING

Meskipun **RandomizedSearchCV** cukup efisien dalam mencari kombinasi hyperparameter optimal, metode ini berisiko **melewatkan parameter penting** karena pencariannya bersifat acak. Oleh karena itu, disarankan untuk menggunakan **GridSearchCV**, karena **GridSearchCV** akan menguji semua kemungkinan kombinasi secara menyeluruh, yang dapat membantu menemukan hasil yang lebih akurat. Namun, yang perlu diperhatikan adalah **GridSearchCV** membutuhkan **waktu komputasi yang jauh lebih lama** dibandingkan dengan **RandomizedSearchCV**, terutama jika jumlah kombinasi yang diuji cukup banyak. Meski demikian, dengan menggunakan **GridSearchCV**, proses pencarian hyperparameter menjadi lebih komprehensif dan memastikan bahwa parameter penting tidak terlewatkan.
